```python
#GRAFICOS DE VENTAS POR CATEGORIA DE CADA TIENDA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Asegúrate de que estas variables contienen los resultados del análisis de ventas por categoría
# vendas_por_categoria_loja1, vendas_por_categoria_loja2, vendas_por_categoria_loja3, vendas_por_categoria_loja4

# Lista de los datos de ventas por categoría para cada tienda
data_categorias = [vendas_por_categoria_loja1, vendas_por_categoria_loja2, vendas_por_categoria_loja3, vendas_por_categoria_loja4]
nombres_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Obtener todas las categorías únicas
all_categories = set()
for data in data_categorias:
    all_categories.update(data.index)
all_categories = sorted(list(all_categories))

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 7))

# Ancho de las barras
bar_width = 0.2

# Posiciones de las barras para cada tienda
x = np.arange(len(all_categories))
positions = [x - 1.5*bar_width, x - 0.5*bar_width, x + 0.5*bar_width, x + 1.5*bar_width]

# Colores para cada tienda
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

# Crear las barras para cada tienda
for i, data in enumerate(data_categorias):
    # Crear un diccionario para acceder fácilmente a las ventas por categoría de la tienda actual
    data_dict = data.to_dict()
    # Obtener la cantidad vendida para cada categoría, o 0 si la categoría no está presente
    sales_counts = [data_dict.get(cat, 0) for cat in all_categories]
    ax.bar(positions[i], sales_counts, bar_width, label=nombres_lojas[i], color=colors[i])

# Añadir etiquetas y título
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Cantidad Vendida')
ax.set_title('Ventas por Categoría en Diferentes Tiendas')
ax.set_xticks(x)
ax.set_xticklabels(all_categories, rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# Asumiendo que los DataFrames loja, loja2, loja3, loja4 ya están cargados

# Lista de Puntos Geográficos por Tienda
geographical_data = {}

# Extraer las coordenadas para cada tienda
geographical_data['Loja 1'] = list(zip(loja['lat'], loja['lon']))
geographical_data['Loja 2'] = list(zip(loja2['lat'], loja2['lon']))
geographical_data['Loja 3'] = list(zip(loja3['lat'], loja3['lon']))
geographical_data['Loja 4'] = list(zip(loja4['lat'], loja4['lon']))

# Función para crear un mapa de calor
def create_heatmap(coordinates, loja_nome):
    if coordinates:
        # Calcular la latitud y longitud promedio para centrar el mapa
        mean_lat = sum([coord[0] for coord in coordinates]) / len(coordinates)
        mean_lon = sum([coord[1] for coord in coordinates]) / len(coordinates)

        # Crear el mapa
        m = folium.Map(location=[mean_lat, mean_lon], zoom_start=10)

        # Añadir el mapa de calor
        HeatMap(coordinates).add_to(m)

        # Guardar el mapa como un archivo HTML
        filepath = f'heatmap_distribucion_{loja_nome}.html'
        m.save(filepath)
        print(f'Mapa de calor para {loja_nome} guardado en {filepath}')
        return filepath
    else:
        print(f"No hay datos de coordenadas para {loja_nome} para generar el mapa de calor.")
        return None

# Crear mapas de calor para cada tienda
heatmap_files = {}
for loja_nome, coords in geographical_data.items():
    filepath = create_heatmap(coords, loja_nome)
    if filepath:
        heatmap_files[loja_nome] = filepath

print("\nMapas de calor generados. Puedes descargarlos desde la sección de archivos de Colab (a la izquierda) y abrirlos en tu navegador web para visualizarlos.")
```
